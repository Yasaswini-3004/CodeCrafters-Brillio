```python
import os
import comtypes.client

# Path to the Resumes folder containing DOCX files
resumes_folder = 'C:\\Users\\desuy\\Desktop\\CodeCrafters\\Frontend\\public\\Resumes'

# Path to the folder where you want to save the converted PDF files
pdf_folder = 'C:\\Users\\desuy\\Desktop\\CodeCrafters\\Frontend\\public\\ResumePDFs'

# Ensure the PDF folder exists, create it if not
if not os.path.exists(pdf_folder):
    os.makedirs(pdf_folder)

# Initialize Word application
word = comtypes.client.CreateObject("Word.Application")
word.Visible = False

# Get all the DOCX files in the Resumes folder
docx_files = [file for file in os.listdir(resumes_folder) if file.endswith('.docx')]

for docx_file in docx_files:
    docx_path = os.path.join(resumes_folder, docx_file)
    pdf_path = os.path.join(pdf_folder, os.path.splitext(docx_file)[0] + '.pdf')

    # Load the DOCX file
    doc = word.Documents.Open(docx_path)
    
    # Save as PDF
    doc.SaveAs(pdf_path, FileFormat=17)  # 17 represents PDF format
    
    # Close the document
    doc.Close()

    print(f"Converted {docx_file} to PDF")

# Quit Word application
word.Quit()
```
